```python
import requests
from bs4 import BeautifulSoup
import time

def extract_article_urls(html_content):
    base_url = "https://www.shfe.com.cn/publicnotice/spotlight/"
    soup = BeautifulSoup(html_content, 'html.parser')
    urls = []
    
    for item in soup.select('div.table_item_info'):
        link = item.select_one('div.info_item_title a')
        if link and link.get('href'):
            relative_path = link['href'].replace('./', '')
            full_url = f"{base_url}{relative_path}"
            urls.append(full_url)
    
    return urls

def crawl_all_pages():
    # 请求配置
    cookies = {'TrsAccessMonitor': 'TrsAccessMonitor-1747408917000-1406419265'}
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36',
        'Referer': 'https://www.shfe.com.cn/publicnotice/'
    }
    
    base_url = "https://www.shfe.com.cn/publicnotice/spotlight/"
    all_articles = []
    page_count = 1  # 页码计数器
    
    # 先爬取第一页（无页码的初始页面）
    # response = requests.get(base_url, headers=headers, cookies=cookies)
    # if response.status_code == 200:
    #     urls = extract_article_urls(response.text)
    #     all_articles.extend(urls)
    #     print(f"初始页面 爬取到 {len(urls)} 条链接")
    # else:
    #     print("初始页面请求失败")
    #     return []
    
    # 爬取后续分页（index_1.html, index_2.html...）
    while True:
        page_url = f"{base_url}index_{page_count}.html"
        response = requests.get(page_url, headers=headers, cookies=cookies)
        
        # 终止条件：页面不存在或无数据
        if response.status_code != 200:
            print(f"页码 {page_count} 页面不存在，停止爬取")
            break
            
        urls = extract_article_urls(response.text)
        if not urls:
            print(f"页码 {page_count} 无数据，停止爬取")
            break
            
        all_articles.extend(urls)
        print(f"页码 {page_count} 爬取到 {len(urls)} 条链接")
        
        page_count += 1
        time.sleep(1)  # 添加请求间隔避免被封
    
    return all_articles

# 执行爬取
if __name__ == "__main__":
    articles = crawl_all_pages()
    print("\n最终爬取结果：")
    print(f"共获取到 {len(articles)} 篇文章链接")
    for url in articles:
        print(url)
```

```python
from bs4 import BeautifulSoup
import re
def parse_article_page(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # 提取标题
    title = soup.select_one('h1.title').get_text(strip=True) if soup.select_one('h1.title') else None
    
    # 提取发布日期
    publish_date = None
    date_span = soup.select_one('.notice_nav .nav_left span')
    if date_span:
        date_text = date_span.get_text(strip=True).replace('发布日期：', '')
        publish_date = date_text.strip()
    
    # 提取文章内容
    content = []
    content_div = soup.select_one('.TRS_Editor')
    if content_div:
        # 处理不同内容元素
        for element in content_div.find_all(['p', 'div', 'span', 'font']):
            # 去除图片标签
            if element.find('img'):
                continue
            # 处理文本内容
            text = element.get_text(strip=True)
            if text:
                content.append(text)
    full_content = ''.join(content)
    
    soup = BeautifulSoup(html_content, 'html.parser')
    content_div = soup.find('div', class_='notice_contain')
    # print(content_div)
    if content_div:
        # 提取文本并清理多余空白
        article_content = content_div.get_text(separator='\n', strip=True)
        full_content=article_content
    else:
        print("未找到内容部分")
    full_content = re.sub(r'\n', '', full_content)
    # 合并内容段落
    
    
    return {
        'title': title,
        'publish_date': publish_date,
        'content': full_content
    }

# 使用示例
article_info = parse_article_page(response.text)
print(f"标题：{article_info['title']}")
print(f"发布日期：{article_info['publish_date']}")
print(f"内容长度：{len(article_info['content'])} 字符")
print("内容片段：", article_info['content'][:200] + "...")
```
